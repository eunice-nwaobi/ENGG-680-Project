```python
import zipfile
import pandas as pd
import os
preprefix = "C:/Users/eenwa/OneDrive - University of Calgary/"
prefix = f"{preprefix}ENGG-680-Project/data/" # location of data file cloned from github (change as needed)
prefixTrain = prefix + "train/"
prefixValid = prefix + "valid/"
```

```python
df = pd.read_excel(f"{prefix}Combined excel.xlsx", usecols="A,B,C,D,H,U,V")
df = df.set_index("Filename")
print(df.head())
filenames=set(df.index)
print(len(filenames))
```

```python
with zipfile.ZipFile(f"{prefixValid}valid00.zip", "r") as zipObj:
  filenames08=set(zipObj.namelist())
  unionFilenames = filenames & filenames08 # intersection symbol
  print(f"The number of files to download from zip is {len(unionFilenames)}")

  for ufn in unionFilenames:
    # rename as name_VA_RB FS Defect
    newname = ufn[:-4] + "_"
    newname += str(df.loc[ufn, "VA"]) + "_"
    newname += str(df.loc[ufn, "RB"])
    newname += str(df.loc[ufn, "FS"])
    newname += str(df.loc[ufn, "Defect"]) + ".png"
    zipObj.getinfo(f"{ufn}").filename = newname
    zipObj.extract(ufn, path=f"{preprefix}/ENGG-680-ProjectDataDestination/")
  zipObj.close()
```
